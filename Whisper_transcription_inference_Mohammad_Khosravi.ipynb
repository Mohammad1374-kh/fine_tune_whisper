# Whisper large persian
This model is a fine-tuned version of whisper large v3 vaialable at [MohammadKhosravi/whisper-large-v3-Persian](https://huggingface.co/MohammadKhosravi/whisper-large-v3-Persian) on the [mozilla-foundation/common_voice_11_0 fa dataset Persian subset](https://huggingface.co/datasets/mozilla-foundation/common_voice_11_0/viewer/fa). It achieved the following results on the evaluation set:

* Loss: 0.2444
* Wer: 26.8658

```python
!pip install -U transformers datasets
!pip install -U librosa  # For audio processing
```

```python
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq
import torch
import librosa
```

# Load the processor and the pre-trained model

```python
processor = AutoProcessor.from_pretrained("MohammadKhosravi/whisper-large-v3-Persian")
model = AutoModelForSpeechSeq2Seq.from_pretrained("MohammadKhosravi/whisper-large-v3-Persian")
```

# Path to the audio file

```python
audio_file_path = "/content/joz az kol.wav"

# Load and process the audio file
# Whisper expects audio to be sampled at 16kHz
speech_array, sampling_rate = librosa.load(audio_file_path, sr=16000)

# Tokenize the input audio
inputs = processor(speech_array, sampling_rate=16000, return_tensors="pt").input_features

# Perform inference
with torch.no_grad():
    predicted_ids = model.generate(inputs)
```

```python
# Decode the prediction
transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]

print("\n--- Transcription ---\n")
print(transcription)
```

```python
from jiwer import wer


# Expected (ground truth) transcription
expected_transcription = "ناامیدی ته نداره!تو هیچ وقت به ته ناامیدی نمیرسی برای همینه هیچ وقت خودکشی نمیکنی!اگر به جاودانگی باورداشته باشی میتونی خودتو خلاص کنی!تا وقتی وحشت از زندگیت نرفته نمیفهمی ترس چقدر ازت زمان میگیره!هیچی مثل احساس گناه آدمو نمیخوره!"

# Calculate WER
error_rate = wer(expected_transcription, transcription)

# Print the WER
print(f"Word Error Rate (WER): {error_rate * 100:.2f}%")
```

As you can see the eror rate is about 5 percent meaing the error rate is the same and 95 percent of the transcription is correct.
